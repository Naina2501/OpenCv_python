filter color

```python
import cv2
import numpy as np

cap=cv2.VideoCapture(0)


def hello(x):
    pass

cv2.namedWindow("demo")

# HSV Trackbars
cv2.createTrackbar("lh","demo",0,255,hello)
cv2.createTrackbar("ls","demo",0,255,hello)
cv2.createTrackbar("lv","demo",0,255,hello)

cv2.createTrackbar("uh","demo",255,255,hello)
cv2.createTrackbar("us","demo",255,255,hello)
cv2.createTrackbar("uv","demo",255,255,hello)

while cap.isOpened():
    r,frame=cap.read()
    if  r==True:
        img=cv2.resize(frame,(400,300))
        hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        
    
        lh = cv2.getTrackbarPos("lh","demo")
        ls = cv2.getTrackbarPos("ls","demo")
        lv = cv2.getTrackbarPos("lv","demo")
    
        uh = cv2.getTrackbarPos("uh","demo")
        us = cv2.getTrackbarPos("us","demo")
        uv = cv2.getTrackbarPos("uv","demo")
    
        lower = np.array([lh, ls, lv])
        upper = np.array([uh, us, uv])
    
        mask = cv2.inRange(hsv_img, lower, upper)
        res = cv2.bitwise_and(img, img, mask=mask)
    
        cv2.imshow("Original", img)
        cv2.imshow("HSV", hsv_img)
        cv2.imshow("Mask", mask)
        cv2.imshow("Result", res)
    
        if cv2.waitKey(1) & 0xFF == ord('p'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

Histogram equalization

CLAHE

```python
import cv2
import numpy as np

# Read image
img = cv2.imread("project_image/back.png")
img = cv2.resize(img, (300, 300))

# Convert to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Define GREEN range (tweak if needed)
lower_green = np.array([35, 40, 40])
upper_green = np.array([85, 255, 255])

# Mask for green background
green_mask = cv2.inRange(hsv, lower_green, upper_green)

# Invert mask  object becomes white
object_mask = cv2.bitwise_not(green_mask)

# Clean mask
kernel = np.ones((5,5), np.uint8)
object_mask = cv2.morphologyEx(object_mask, cv2.MORPH_OPEN, kernel)


# Create white background
white_bg = np.ones_like(img, dtype=np.uint8) * 255

# Extract object

object_only = cv2.bitwise_and(img, img, mask=object_mask)

# Extract white background
background = cv2.bitwise_and(white_bg, white_bg, mask=green_mask)

# Combine object + white background
final = cv2.add(object_only, background)

# Show results
cv2.imshow("Original", img)
cv2.imshow("Final Output", final)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Read and resize background image
img = cv2.imread("project_image/back.png")
img = cv2.resize(img, (600, 500))
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Read and convert reference image
img1 = cv2.imread("project_image/copy.png")
hsv2 = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)

# Calculate histogram of reference image
hist = cv2.calcHist([hsv2], [0, 1], None, [180, 256], [0, 180, 0, 256])
cv2.normalize(hist, hist, 0, 255, cv2.NORM_MINMAX)

# Back projection
mask = cv2.calcBackProject([hsv], [0, 1], hist, [0, 180, 0, 256], 1)

# Smoothing
ker = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
mask = cv2.filter2D(mask, -1, ker)

# Thresholding
_, thr = cv2.threshold(mask, 200, 255, cv2.THRESH_BINARY)

# Convert to 3-channel mask
thr = cv2.merge((thr, thr, thr))

# Apply mask
res = cv2.bitwise_and(img, thr)

# Display
cv2.imshow("Result", res)
cv2.imshow("Original", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

# Read first frame
ret, frame = cap.read()
if not ret:
    exit()

# Select ROI (object to track)
x, y, w, h = cv2.selectROI("Select Object", frame, False)
roi = frame[y:y+h, x:x+w]

# Convert ROI to HSV
hsv_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

# Create mask to ignore low light
mask = cv2.inRange(hsv_roi,
                   np.array((0., 60., 32.)),
                   np.array((180., 255., 255.)))

# Histogram of ROI
roi_hist = cv2.calcHist([hsv_roi], [0], mask, [180], [0, 180])
cv2.normalize(roi_hist, roi_hist, 0, 255, cv2.NORM_MINMAX)

# Initial tracking window
track_window = (x, y, w, h)

# Termination criteria
term_crit = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 1)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Back projection
    dst = cv2.calcBackProject([hsv], [0], roi_hist, [0, 180], 1)

    # Apply CamShift
    ret, track_window = cv2.CamShift(dst, track_window, term_crit)

    # Draw tracking box
    pts = cv2.boxPoints(ret)
    pts = np.int32(pts)
    cv2.polylines(frame, [pts], True, (0, 255, 0), 2)

    cv2.imshow("Object Tracking", frame)

    if cv2.waitKey(1) & 0xFF == ord("p"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread("project_image/Sudoku.jpg")
img=cv2.resize(img,(400,400))
gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

edg = cv2.Canny(gry,20,250)

lines = cv2.HoughLines(edg,1,np.pi/180,200)
for r,th in lines[0]:
    a = np.cos(th)
    b = np.sin(th)
    x0=a*r
    y0=b*r
    x1 = int(x0+1000*(-b))
    y1 = int(y0+1000*(a))
    x2 = int(x0-1000*(-b))
    y2 = int(y0-1000*(a))
    
    cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)
    
       
       

cv2.imshow("original",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
